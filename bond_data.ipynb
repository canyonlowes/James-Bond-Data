```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('james_bond_data_cleansed.csv')
print (data)
```

```python
plt.scatter(data['imdb'], data['rotten_tomatoes'])
plt.title('Scatter Plot of Ratings')
plt.xlabel('Average IMDB Rating')
plt.ylabel('Average Rotten Tomatoes Rating')
```

```python
plt.scatter(data['imdb'], data['bond_kills'])
plt.title('Bond kills vs IMDB Rating')
plt.xlabel('IMDB Rating')
plt.ylabel('Bond Kills')
```

```python
x = data['release_year']
y = data['income_world']
y1 = data['movie_budget']

plt.plot(x,y1, 'b-o')
plt.plot(x,y,'r-o')
plt.xlabel('Release Year')
plt.legend()
plt.show()
```

```python
x = data['release_year']
y = data['income_world'] / data['movie_budget']

plt.bar(x,y, label='Income/budget ratio')
plt.xlabel('Release Year')
plt.ylabel('income/budget ratio')
plt.title('income to budget ratio by release year')
plt.legend()
plt.show()
```
